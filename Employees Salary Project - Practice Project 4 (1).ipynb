# Taking Weight Column as the target column

```python
x=df_new1.iloc[:,[0,1,3,4,5,6,7,8,9,10]]
x
```

```python
y=pd.DataFrame(df_new1.iloc[:,2])
y
```

Finding Best Random State

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
best_r_state=0
accu=0

for i in range(30,200):
    train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3,random_state=i)
    mod=LinearRegression()
    mod.fit(train_x,train_y)
    pred_y=mod.predict(test_x)
    tempaccu=r2_score(test_y,pred_y)
    if tempaccu>accu:
        accu=tempaccu
        best_r_state=i

print(f"Best random_state {best_r_state}")
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3,random_state=34)
lm=LinearRegression()
lm.fit(train_x,train_y)
lm_y=lm.predict(test_x)

from sklearn.metrics import r2_score
print(r2_score(test_y,lm_y))
```

Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB

mnb=MultinomialNB()
mnb.fit(train_x,train_y)
mnb_y=mnb.predict(test_x)

print(r2_score(test_y,mnb_y))
```

KNeighborsRegressor

```python
from sklearn.neighbors import KNeighborsRegressor

knn=KNeighborsRegressor(n_neighbors=5)
knn.fit(train_x,train_y)
knn_y=knn.predict(test_x)

print(r2_score(test_y,knn_y))
```

RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor

rfr=RandomForestRegressor(n_estimators=100,random_state=95)
rfr.fit(train_x,train_y)
rfr_y=rfr.predict(test_x)

print(r2_score(test_y,rfr_y))
```

AdaBoostClassifier

```python
from sklearn.ensemble import AdaBoostRegressor

abr=AdaBoostRegressor()
abr.fit(train_x,train_y)
abr_y=abr.predict(test_x)

print(r2_score(test_y,abr_y))
```

GradientBoostingClassifier

```python
from sklearn.ensemble import GradientBoostingRegressor

gbc=GradientBoostingRegressor()
gbc.fit(train_x,train_y)
gbc_y=gbc.predict(test_x)

print(r2_score(test_y,gbc_y))
```

GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR

svr=SVR()
parameters={'kernel':['linear','rbf','poly'],'C':[1,10]}
clf=GridSearchCV(svr,parameters)
clf.fit(x,y)
```

```python
print(clf.best_params_)
```

```python
svr_rbf=SVR('kerel'='rbf','C'=1)
svr_rbf.fit(train_x,train_y)
svr_y=svr_rbf.predict(test_x)

print(r2_score(test_y,svr_y))
```

On the basis of Accuracy Score The Best Model is done by Linear Regression having accuracy score of 0.55.
